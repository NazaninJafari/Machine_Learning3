```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
covid_data = pd.read_csv('New_folder/COVID-19-time-series-clean-complete.csv')
covid_data
```

```python
covid_data['Date'] = pd.to_datetime(covid_data['Date'])

covid_data['YearMonth'] = covid_data['Date'].map(lambda date: date.year * 100 + date.month )
covid_data
```

```python
april_cases = covid_data.query("YearMonth == 202004")
april_cases
```

```python
sum_aprilCases = april_cases.groupby(['Country/Region'])['New cases'].sum().reset_index()
sum_aprilCases
```

```python
sorted_cases = sum_aprilCases.sort_values('New cases', ascending= False).reset_index()
sorted_cases.head(8)
```

```python
#iranien_covidCases = covid_data[covid_data['Country/Region'] == 'Iran'].reset_index()
#iranien_covidCases = iranien_covidCases.groupby(['YearMonth'])['New cases'].sum()
iranien_covidCases = covid_data.loc[(covid_data['Country/Region'] == 'Iran')].groupby(['YearMonth'])['New cases'].sum().reset_index()
iranien_covidCases
```

```python
iranien_covidDeath = covid_data.loc[(covid_data['Country/Region'] == 'Iran')].groupby(['YearMonth'])['New deaths'].sum().reset_index()
iranien_covidDeath
```

```python
plt.figure(figsize= (20,8))
#plt.plot(iranien_covidCases['YearMonth'].to_numpy(dtype = str), iranien_covidCases['New cases'])
plt.fill_between(iranien_covidCases['YearMonth'].to_numpy(dtype = str) , iranien_covidCases['New cases'], alpha = 0.5)
plt.fill_between(iranien_covidDeath['YearMonth'].to_numpy(dtype = str), iranien_covidDeath['New deaths'])
plt.legend(['Cases' , 'Deaths'])
plt.title('New cases vs New deaths in Iran')
plt.grid()
```
